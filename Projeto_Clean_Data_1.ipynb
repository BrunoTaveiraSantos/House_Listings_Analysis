```python
import pandas as pd
```

```python
Houses_Lisbon = pd.read_excel("Casas_Lisboa.xlsx")
Houses_data_set = pd.read_csv("portugal_houses.csv")
```

```python
Locations_wanted = ["Lisboa","Cascais","Torres Vedras","Amadora","Sintra","Odivelas","Oeiras","Lourinhã","Mafra","Vila Franca de Xira","Loures"]
```

```python
filtered_df = Houses_data_set[Houses_data_set["location"].isin(Locations_wanted)]
```

```python
filtered_df
```

```python
filtered_df = filtered_df.drop(filtered_df.columns[[7, 8,10,11]],axis = 1)
```

```python
#Clean the date to only year 

date = list(map(lambda x: x.split("-"),filtered_df["started_at"]))
clean_date = []

for item in date:
    clean_date.append(item[0])

filtered_df["started_at"] = clean_date
```

```python
#clean the area gross to only number
area = list(map(lambda x: str(x).split("."),filtered_df["area.gross"]))
clean_area = []
for item in area:
    clean_area.append(item[0])

filtered_df["area.gross"] = clean_area
```

```python
import re
regex = r"T\d+"
clean_title = []
clean_title2 = []
for string in filtered_df["title"]:
    numbers = re.findall(regex, string)
    clean_title.append(numbers)
clean_title
filtered_df["title"] = clean_title
```

```python
filtered_houses_lisbon = Houses_Lisbon.drop(Houses_Lisbon.columns[[3]],axis = 1)
```

```python
# Clean Square meters
area_lisbon = list(map(lambda x: str(x).split(" "),filtered_houses_lisbon["Square_meters"]))
clean_area_lisbon = []
for item in area_lisbon:
    clean_area_lisbon.append(item[0])

filtered_houses_lisbon["Square_meters"] = clean_area_lisbon
```

```python
#remove spaces from price 
clean_price = list(map(lambda x: x.replace(" ",""),filtered_houses_lisbon["Price"]))
clean_price = list(map(lambda x: int(x),clean_price))
filtered_houses_lisbon["Price"] = clean_price
```
