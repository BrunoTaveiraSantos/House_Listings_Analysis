```python
def remove_string_from_list(str_list, str_to_remove):
    """
    Removes every occurrence of a given string from a list of strings.
    
    Parameters:
    -----------
    str_list: list
        List of strings from which to remove the given string.
    
    str_to_remove: str
        String to remove from the list.
    
    Returns:
    --------
    new_list: list
        List of strings with every occurrence of the given string removed.
    """
    new_list = [x for x in str_list if x != str_to_remove]
    return new_list
```

```python
def remove_string_from_list2(str_list, str_to_remove):
    """
    Removes every occurrence of a given string from a list of strings.
    
    Parameters:
    -----------
    str_list: list
        List of strings from which to remove the given string.
    
    str_to_remove: str
        String to remove from the list.
    
    Returns:
    --------
    new_list: list
        List of strings with every occurrence of the given string removed.
    """
    substring = "casas para comprar"
    new_list = [x for x in str_list if substring in x]
    return new_list
```
