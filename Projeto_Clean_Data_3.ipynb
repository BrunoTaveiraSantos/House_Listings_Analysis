```python
import pandas as pd
```

```python
Houses_Lisbon_2 = pd.read_excel("Lisbon_House_Final_1.xlsx")
supercasa = pd.read_csv("supercasa2.csv")
```

```python
supercasa 
```

```python
clean_category = list(map(lambda x: x.split(" "),supercasa["Title"]))
```

```python
clean_category = list(map(lambda x: x[0],clean_category))
```

```python
supercasa["Title"] = clean_category
```

```python
# Clean Price 

supercasa["Price"] = list(map(lambda x: x.replace(".",""),supercasa["Price"]))
```

```python
supercasa = supercasa.drop(supercasa[supercasa['Squared Meters'] == 'not specified'].index)
```

```python
supercasa["Year"]= 2023
supercasa["region"]= "Lisboa"
display(supercasa)
```

```python
Houses_Lisbon_2
```

```python
supercasa = supercasa.rename(columns={'Title': 'Category', 'Description': 'Location', 'Rooms': 'Typology', "Squared Meters": "Square_meters"})
```

```python
supercasa = supercasa.reindex(columns=["Location","region","Typology","Square_meters","Price","Category"])
```

```python
supercasa["Year"] = 2023
supercasa
Houses_Lisbon_2
```

```python
Lisbon_House_DF_Final_2 = pd.concat([Houses_Lisbon_2, supercasa], ignore_index=True)
```

```python
Lisbon_House_DF_Final_2["Location"].value_counts()
```

```python
Lisbon_House_DF_Final_2 = Lisbon_House_DF_Final_2.drop(Lisbon_House_DF_Final_2[Lisbon_House_DF_Final_2["Location"] == 'Évora'].index)
```

```python
Lisbon_House_DF_Final_2.to_excel('Lisbon_House_Final_2.xlsx', index=False)
```
