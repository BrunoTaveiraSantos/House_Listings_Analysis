```python
import pandas as pd
```

```python
imovirtual = pd.read_csv("imo_cleaned.csv")
houses_lisbon = pd.read_excel("Lisbon_House.xlsx")
```

```python
display(houses_lisbon)
display(imovirtual)
```

```python
filtered_imovirtual = imovirtual.drop(imovirtual.columns[0],axis = 1)
```

```python
filtered_imovirtual
```

```python
#Clean Area 
```

```python
area_imovirtual = list(map(lambda x: str(x).split(","),filtered_imovirtual["Area"]))
```

```python
area_imovirtual = list(map(lambda x: x[0],area_imovirtual))
```

```python
area_imovirtual = list(map(lambda x: x.replace(" m²",""),area_imovirtual))
```

```python
filtered_imovirtual["Area"] = area_imovirtual
```

```python
#Dropnull_values 
filtered_imovirtual = filtered_imovirtual.dropna()
```

```python

new_name = []

for item in filtered_imovirtual["Name"]:
    if item != "Apartamento":
        new_name.append("Apartamento")
    else:
        new_name.append(item)
```

```python
filtered_imovirtual["Name"] = new_name
```

```python
filtered_imovirtual["Location"] = "Lisboa"
filtered_imovirtual["Region"] = "Lisboa"
```

```python
filtered_imovirtual
filtered_imovirtual = filtered_imovirtual.rename(columns={'Name': 'Category', 'Rooms': 'Typology', 'Area': 'Square_meters'})
```

```python
filtered_imovirtual = filtered_imovirtual.rename(columns={'Region': 'region'})
```

```python
filtered_imovirtual = filtered_imovirtual.reindex(columns=["Location","region","Typology","Square_meters","Price","Category"])
```

```python
filtered_imovirtual["Year"] = 2023
filtered_imovirtual
```

```python
Lisbon_House_DF_Final_1 = pd.concat([houses_lisbon, filtered_imovirtual], ignore_index=True)
```

```python
Lisbon_House_DF_Final_1.to_excel('Lisbon_House_Final_1.xlsx', index=False)
```
